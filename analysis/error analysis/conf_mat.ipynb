```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('CLIP-Full(main).csv', header=None)
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(
    df,
    annot=False,
    cmap="Blues",
    # linewidths=0.02,
    # linecolor='lightgray',
    xticklabels=True,
    yticklabels=True,
    cbar=False
    # cbar_kws={'label': "Frequency"}
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks(rotation=0, fontsize=8)
plt.yticks(fontsize=8)

plt.tight_layout()
plt.show()
```
